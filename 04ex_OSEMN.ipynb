### 2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
R = np.random.rand(5,5)
file = open("data.txt","w")
file.write("\n".join("\t".join(str(r2) for r2 in r1) for r1 in R))
file.close()

file = open("data.txt")
print(file.read())
file.close()
```
